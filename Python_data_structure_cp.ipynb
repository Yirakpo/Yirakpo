```python
def main():
    # Create a list to store the shopping items
    shopping_list = []

    # Define a dictionary to store menu options
    menu_options = {
        '1': 'Add item',
        '2': 'Remove item',
        '3': 'View list',
        '4': 'Exit'
    }

    # Define a set to store unique items (just to demonstrate use of a set)
    unique_items = set()

    # Main loop
    while True:
        # Display menu options
        print("\nMenu:")
        for key, value in menu_options.items():
            print(f"{key}: {value}")

        # Get user selection
        choice = input("Enter your choice: ")

        # Process user selection
        if choice == '1':
            if len(shopping_list) < 10:  # Limiting the number of items to 10
                item = input("Enter an item to add: ")
                shopping_list.append(item)
                unique_items.add(item)  # Add item to the set
                print(f"'{item}' has been added to the shopping list.")
            else:
                print("Shopping list is full. You can't add more than 10 items.")

        elif choice == '2':
            item = input("Enter an item to remove: ")
            if item in shopping_list:
                shopping_list.remove(item)
                unique_items.discard(item)  # Remove item from the set
                print(f"'{item}' has been removed from the shopping list.")
            else:
                print(f"'{item}' is not in the shopping list.")

        elif choice == '3':
            if shopping_list:
                print("Current shopping list:")
                for i in range(len(shopping_list)):
                    print(f"{i + 1}: {shopping_list[i]}")
            else:
                print("The shopping list is empty.")

        elif choice == '4':
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
